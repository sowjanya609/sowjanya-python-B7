```python
# Question2


def isPrime(n):
    if (n<2):
        return False
    for i in range(2,n//2+1):
        if(n%i==0):
            return False
        return True
a = int(input("Enter lower range :"))
b = int(input("Enter higher range :"))
for x in range(a,b):
    if isPrime(x):
        print(x, end = ' ')
```

```python
lst = list(range(2500))
```

```python
print(lst)
```

```python
lst_Prime = []
```

```python
for item in lst:
    if isPrime(item):
        lst_Prime.append(item)
```

```python
print(lst_Prime)
```

```python
lst_Prime_one = filter(isPrime,lst)
```

```python
list(list(lst_Prime_one))
```
